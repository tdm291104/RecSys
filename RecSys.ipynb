```python
import numpy as np
import pandas as pd
from scipy import sparse
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from typing import List, Dict, Optional, Tuple
import json
```

```python
import pandas as pd

class DataStore:
    def __init__(self, recipes_csv: str, ratings_csv: str):
        # ƒê·ªçc d·ªØ li·ªáu
        self.recipes = pd.read_csv(recipes_csv, encoding="utf-8")
        self.ratings = pd.read_csv(ratings_csv, encoding="utf-8")

        # Chu·∫©n h√≥a t√™n c·ªôt
        self.recipes.columns = [c.strip().lower() for c in self.recipes.columns]
        self.ratings.columns = [c.strip().lower() for c in self.ratings.columns]

        # ƒê·∫£m b·∫£o c·ªôt food_id t·ªìn t·∫°i
        assert 'id' in self.recipes.columns or 'food_id' in self.recipes.columns, \
            "recipes.csv must have 'id' or 'food_id'"
        if 'food_id' not in self.recipes.columns:
            self.recipes.rename(columns={'id': 'food_id'}, inplace=True)

        # ƒê·∫£m b·∫£o ratings.csv c√≥ c√°c c·ªôt b·∫Øt bu·ªôc
        for col in ['user_id', 'food_id', 'rating']:
            assert col in self.ratings.columns, f"ratings.csv must have column '{col}'"

        # ƒêi·ªÅn gi√° tr·ªã r·ªóng cho c√°c c·ªôt text
        for col in ['dish_name','description','cooking_method','ingredients','dish_type']:
            if col in self.recipes.columns:
                self.recipes[col] = self.recipes[col].fillna('')

        # Chuy·ªÉn c·ªôt s·ªë th√†nh numeric
        for col in ['serving_size', 'cooking_time', 'calories']:
            if col in self.recipes.columns:
                self.recipes[col] = pd.to_numeric(self.recipes[col], errors='coerce')

        # T·∫°o mapping cho recipe_id
        self.recipe_ids = self.recipes['food_id'].astype(int).tolist()
        self.recipe_id_to_idx = {rid: i for i, rid in enumerate(self.recipe_ids)}

        # L·ªçc ratings theo food_id h·ª£p l·ªá
        valid_food_ids = set(self.recipe_ids)
        self.ratings = self.ratings[self.ratings['food_id'].isin(valid_food_ids)].copy()

        # T·∫°o mapping cho user_id
        self.user_ids = sorted(self.ratings['user_id'].unique().tolist())
        self.user_id_to_idx = {uid: i for i, uid in enumerate(self.user_ids)}

        # In th√¥ng tin d·ªØ li·ªáu
        print("üìä Data loaded successfully!")
        print(f"   - Recipes: {len(self.recipes)} items")
        print(f"   - Ratings: {len(self.ratings)} ratings")
        print(f"   - Users: {len(self.user_ids)} unique users")
        print(f"   - Recipe columns: {list(self.recipes.columns)}")
```

```python
RECIPES_CSV = "foods.csv"  # Change this path
RATINGS_CSV = "ratings.csv"  # Change this path

data = DataStore(RECIPES_CSV, RATINGS_CSV)
```

```python
import pandas as pd
import numpy as np
from typing import List, Tuple
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

class ContentModel:
    def __init__(self, recipes: pd.DataFrame):
        self.recipes = recipes
        self.vectorizer = TfidfVectorizer(
            ngram_range=(1, 2),
            min_df=2,
            max_features=50000
        )
        self.tfidf = None  # sparse matrix [n_items, n_feats]

    def _compose_text(self, row: pd.Series) -> str:
        """Gh√©p c√°c c·ªôt text th√†nh 1 chu·ªói duy nh·∫•t"""
        parts = []
        for col in ['ingredients', 'dish_name', 'description', 'cooking_method']:
            if col in row and isinstance(row[col], str):
                parts.append(row[col].lower())
        return ' . '.join(parts)

    def fit(self):
        """Hu·∫•n luy·ªán TF-IDF cho to√†n b·ªô recipes"""
        texts = self.recipes.apply(self._compose_text, axis=1).tolist()
        self.tfidf = self.vectorizer.fit_transform(texts)
        print(f"üîß Content model fitted:")
        print(f"   - TF-IDF shape: {self.tfidf.shape}")
        print(f"   - Vocabulary size: {len(self.vectorizer.vocabulary_)}")
        return self

    def query_from_ingredients(self, ingredients: List[str], topk: int = 50) -> Tuple[np.ndarray, np.ndarray]:
        """Tr·∫£ v·ªÅ (indices, scores) c·ªßa top-k m√≥n ƒÉn g·∫ßn nh·∫•t theo nguy√™n li·ªáu"""
        if not ingredients:
            return np.array([]), np.array([])
        query = ' '.join(ingredients).lower()
        qvec = self.vectorizer.transform([query])
        sims = cosine_similarity(qvec, self.tfidf).ravel()
        idx = np.argpartition(-sims, range(min(topk, len(sims))))[:topk]
        order = np.argsort(-sims[idx])
        return idx[order], sims[idx][order]
    
    def recommend_similar_items(self, food_id: int, topk: int = 10) -> Tuple[np.ndarray, np.ndarray]:
        """
        G·ª£i √Ω c√°c m√≥n ƒÉn t∆∞∆°ng t·ª± v·ªõi 1 m√≥n c·ª• th·ªÉ (d·ª±a v√†o food_id).
        Tr·∫£ v·ªÅ: (indices, similarity_scores)
        """
        # T√¨m index c·ªßa m√≥n ƒÉn trong dataframe
        if 'food_id' in self.recipes.columns:
            try:
                item_idx = self.recipes.index[self.recipes['food_id'] == food_id][0]
            except IndexError:
                raise ValueError(f"‚ö†Ô∏è Food ID {food_id} kh√¥ng t·ªìn t·∫°i trong recipes.")
        else:
            # N·∫øu dataframe kh√¥ng c√≥ c·ªôt food_id, gi·∫£ s·ª≠ index ch√≠nh l√† food_id
            item_idx = food_id
        
        # L·∫•y vector TF-IDF c·ªßa m√≥n ƒÉn n√†y
        item_vec = self.tfidf[item_idx]
        sims = cosine_similarity(item_vec, self.tfidf).ravel()

        # L·∫•y top-k m√≥n t∆∞∆°ng t·ª± (b·ªè ch√≠nh n√≥ ra)
        sims[item_idx] = -1  # Lo·∫°i b·ªè ch√≠nh m√≥n n√†y kh·ªèi g·ª£i √Ω
        idx = np.argpartition(-sims, range(min(topk, len(sims))))[:topk]
        order = np.argsort(-sims[idx])

        return idx[order], sims[idx][order]

    
if 'data' in locals():
    print("üß™ Testing Content-Based Model...")
    
    # Create and fit the model
    content_model = ContentModel(data.recipes)
    content_model.fit()
    
    # Test with Vietnamese ingredients
    test_ingredients = ["th·ªãt b√≤", "h√†nh l√°", "n∆∞·ªõc m·∫Øm"]
    print(f"\nüîç Searching for recipes with: {test_ingredients}")
    
    indices, scores = content_model.query_from_ingredients(test_ingredients, topk=5)
    
    print("\nüìã Top content-based recommendations:")
    for i, (idx, score) in enumerate(zip(indices, scores)):
        recipe = data.recipes.iloc[idx]
        dish_name = recipe.get('dish_name', 'N/A')
        print(f"{i+1}. {dish_name} (Score: {score:.4f})")
else:
    print("‚ùå Please load data first in the previous cell")
```

```python
class CollabModel:
    def __init__(self, ratings: pd.DataFrame, recipe_id_to_idx: Dict[int,int], user_id_to_idx: Dict[int,int]):
        self.ratings = ratings
        self.recipe_id_to_idx = recipe_id_to_idx
        self.user_id_to_idx = user_id_to_idx
        self.R = None              # user-item CSR matrix
        self.item_cosine = None    # item-item similarity (sparse)

    def fit(self, min_interactions_per_item: int = 2):
        """Build user-item matrix and compute item-item similarities"""
        # Map ids to indices
        valid = self.ratings[self.ratings['food_id'].isin(self.recipe_id_to_idx.keys()) &
                            self.ratings['user_id'].isin(self.user_id_to_idx.keys())]
        ui = valid['user_id'].map(self.user_id_to_idx).astype(int).to_numpy()
        ii = valid['food_id'].map(self.recipe_id_to_idx).astype(int).to_numpy()
        vv = valid['rating'].astype(float).to_numpy()

        n_users = len(self.user_id_to_idx)
        n_items = len(self.recipe_id_to_idx)
        self.R = sparse.coo_matrix((vv, (ui, ii)), shape=(n_users, n_items)).tocsr()

        # Normalize by item mean to reduce user scale bias
        item_means = np.asarray(self.R.mean(axis=0)).ravel()
        R_centered = self.R - sparse.csr_matrix(np.tile(item_means, (self.R.shape[0], 1)))

        # Compute item-item cosine similarity
        print("üîß Computing item-item similarities...")
        sims = cosine_similarity(R_centered.T, dense_output=False)
        sims.setdiag(0.0)
        self.item_cosine = sims.tocsr()
        
        print(f"‚úÖ Collaborative model fitted:")
        print(f"   - User-item matrix shape: {self.R.shape}")
        print(f"   - Valid ratings: {len(valid)}")
        print(f"   - Sparsity: {(1 - self.R.nnz / (n_users * n_items)) * 100:.2f}%")
        return self

    def recommend_for_user(self, user_id: int, topk: int = 50) -> Tuple[np.ndarray, np.ndarray]:
        if user_id not in self.user_id_to_idx:
            return np.array([], dtype=int), np.array([])
            
        uidx = self.user_id_to_idx[user_id]
        user_row = self.R.getrow(uidx)
        
        # T√≠nh ƒëi·ªÉm g·ª£i √Ω
        scores = user_row.dot(self.item_cosine).toarray().ravel()
        rated_items = user_row.indices
        scores[rated_items] = -np.inf
        
        # L·∫•y top-k ch·ªâ s·ªë
        top = np.argpartition(-scores, range(min(topk, np.isfinite(scores).sum())))[:topk]
        order = np.argsort(-scores[top])
        top_indices = top[order]
        top_scores = scores[top][order]
        
        # ƒê·ªïi t·ª´ index -> food_id
        idx_to_recipe_id = {v: k for k, v in self.recipe_id_to_idx.items()}
        top_food_ids = np.array([idx_to_recipe_id[idx] for idx in top_indices], dtype=int)
        
        return top_food_ids, top_scores

if 'data' in locals():
    print("üß™ Testing Collaborative Filtering Model...")
    
    # Create and fit the model
    collab_model = CollabModel(data.ratings, data.recipe_id_to_idx, data.user_id_to_idx)
    collab_model.fit()
    
    # Test with a random user
    if len(data.user_ids) > 0:
        test_user = data.user_ids[0]  # Take first user
        print(f"\nüë§ Getting recommendations for User {test_user}")
        
        # Show user's rating history first
        user_ratings = data.ratings[data.ratings['user_id'] == test_user].sort_values('rating', ascending=False)
        print(f"\nüìä User {test_user}'s rating history (top 5):")
        for _, row in user_ratings.head().iterrows():
            recipe = data.recipes[data.recipes['food_id'] == row['food_id']]
            if not recipe.empty:
                dish_name = recipe.iloc[0].get('dish_name', 'N/A')
                print(f"   - {dish_name}: {row['rating']}/5")
        
        # Get collaborative recommendations
        indices, scores = collab_model.recommend_for_user(test_user, topk=5)
        
        print(f"\nüìã Top collaborative recommendations:")
        for i, (idx, score) in enumerate(zip(indices, scores)):
            recipe = data.recipes.iloc[idx]
            dish_name = recipe.get('dish_name', 'N/A')
            print(f"{i+1}. {dish_name} (Score: {score:.4f})")
    else:
        print("‚ùå No users found in the dataset")
else:
    print("‚ùå Please load data first")
```

```python
class HybridRecommender:
    def __init__(self, data):
        self.data = data
        print("üîß Building content-based model...")
        self.content = ContentModel(data.recipes).fit()
        print("üîß Building collaborative filtering model...")
        self.collab = CollabModel(data.ratings, data.recipe_id_to_idx, data.user_id_to_idx).fit()
        print("‚úÖ Hybrid recommender initialized!")

    def _apply_filters(self, candidates: pd.DataFrame, filters: Dict) -> pd.DataFrame:
        """Apply requirement filters"""
        df = candidates
        if not filters:
            return df
            
        print(f"üîç Applying filters: {filters}")
        original_count = len(df)
        
        if 'dish_type' in filters and 'dish_type' in df.columns and filters['dish_type']:
            ft = filters['dish_type']
            if isinstance(ft, (list, tuple, set)):
                df = df[df['dish_type'].isin(ft)]
            else:
                df = df[df['dish_type'] == ft]
                
        if 'serving_size' in filters and 'serving_size' in df.columns and filters['serving_size'] is not None:
            df = df[df['serving_size'] == filters['serving_size']]
            
        if 'max_cooking_time' in filters and 'cooking_time' in df.columns and filters['max_cooking_time'] is not None:
            df = df[df['cooking_time'] <= float(filters['max_cooking_time'])]
            
        if 'max_calories' in filters and 'calories' in df.columns and filters['max_calories'] is not None:
            df = df[df['calories'] <= float(filters['max_calories'])]
            
        print(f"   Filtered from {original_count} to {len(df)} items")
        return df

    def recommend(
        self,
        basis: str = 'hybrid',
        user_id: Optional[int] = None,
        detected_ingredients: Optional[List[str]] = None,
        k: int = 10,
        filters: Optional[Dict] = None,
        alpha: float = 0.6,
    ) -> List[Dict]:
        """
        Return a ranked list of recipes.
        basis: 'content' | 'user' | 'hybrid'
        alpha: weight for content score when basis='hybrid'
        """
        basis = basis.lower()
        print(f"\nüéØ Generating recommendations with basis='{basis}', k={k}, alpha={alpha}")
        
        recipes_df = self.data.recipes.copy()

        # --- Content scores ---
        content_idx = np.array([], dtype=int)
        content_scores = np.array([])
        if basis in ('content','hybrid'):
            print("üìö Computing content-based scores...")
            # If ingredients not provided, build a query from user's highly rated items (if any)
            if (not detected_ingredients) and (user_id is not None):
                # use top-N items rated by user as pseudo-ingredients (from titles)
                u = self.data.user_id_to_idx.get(user_id, None)
                if u is not None:
                    # get items user liked most
                    ur = self.collab.R.getrow(u)
                    liked = ur.toarray().ravel()
                    top_items = liked.argsort()[::-1][:5]
                    pseudo_terms = []
                    for idx in top_items:
                        name = recipes_df.iloc[idx]['dish_name'] if 'dish_name' in recipes_df.columns else ''
                        pseudo_terms.append(str(name))
                    detected_ingredients = pseudo_terms or detected_ingredients
                    print(f"   Using user's liked dishes as query: {pseudo_terms[:3]}...")
                    
            if not detected_ingredients:
                # final fallback: use common Vietnamese cooking terms
                detected_ingredients = ['th·ªãt', 'rau', 'n∆∞·ªõc m·∫Øm']
                print(f"   Using default ingredients: {detected_ingredients}")
            else:
                print(f"   Using provided ingredients: {detected_ingredients}")
                
            content_idx, content_scores = self.content.query_from_ingredients(detected_ingredients, topk=max(k*5, 50))
            print(f"   Found {len(content_idx)} content candidates")

        # --- Collaborative scores ---
        collab_idx = np.array([], dtype=int)
        collab_scores = np.array([])
        if basis in ('user','hybrid') and (user_id is not None):
            print(f"üë• Computing collaborative scores for user {user_id}...")
            collab_idx, collab_scores = self.collab.recommend_for_user(user_id, topk=max(k*5, 50))
            print(f"   Found {len(collab_idx)} collaborative candidates")

        # Combine candidates
        cand_indices = set()
        for arr in [content_idx, collab_idx]:
            cand_indices.update(arr.tolist())
            
        if not cand_indices:
            # no info at all ‚Äî return popular items by average rating
            print("üìà Falling back to popularity-based recommendations")
            valid_ids = set(self.data.recipe_id_to_idx.keys())
            pop = pop[pop.index.isin(valid_ids)]
            cand_indices = [self.data.recipe_id_to_idx[rid] for rid in pop.index[:max(k*5, 50)]]
        else:
            cand_indices = list(cand_indices)

        print(f"üîó Total candidates: {len(cand_indices)}")

        # Build score vectors aligned to candidates
        idx_to_rank = {idx: i for i, idx in enumerate(cand_indices)}
        s_content = np.zeros(len(cand_indices), dtype=float)
        s_collab = np.zeros(len(cand_indices), dtype=float)
        
        for i, idx in enumerate(content_idx):
            if idx in idx_to_rank:
                s_content[idx_to_rank[idx]] = content_scores[i]
                
        for i, idx in enumerate(collab_idx):
            if idx in idx_to_rank:
                s_collab[idx_to_rank[idx]] = collab_scores[i]

        # Combine scores based on basis
        if basis == 'content':
            s_final = s_content
            print("üìä Using content-based scores only")
        elif basis == 'user':
            s_final = s_collab
            print("üìä Using collaborative scores only")
        else:
            # normalize to [0,1] before mixing
            def norm(x):
                if np.all(x == 0) or np.all(~np.isfinite(x)):
                    return np.zeros_like(x)
                return (x - np.nanmin(x)) / (np.nanmax(x) - np.nanmin(x) + 1e-8)
            s_final = alpha * norm(s_content) + (1 - alpha) * norm(s_collab)
            print(f"üìä Using hybrid scores (Œ±={alpha} content + {1-alpha} collaborative)")

        # Rank candidates
        order = np.argsort(-s_final)
        ranked_idx = [cand_indices[i] for i in order]
        ranked_df = self.data.recipes.iloc[ranked_idx].copy()
        ranked_df = ranked_df.copy()
        ranked_df.loc[:, '__score'] = s_final[order]

        # Apply filters
        ranked_df = self._apply_filters(ranked_df, filters or {})

        # Take top-k
        top = ranked_df.head(k)
        cols = [c for c in ['food_id','dish_name','dish_type','serving_size','cooking_time','calories','description'] if c in top.columns]
        cols = cols + ['__score']
        out = top[[c for c in [
            'food_id', 'dish_name', 'dish_type', 'serving_size',
            'cooking_time', 'calories', 'description', '__score'
        ] if c in top.columns]].rename(columns={'food_id': 'id', '__score': 'score'})
        
        print(f"‚úÖ Generated {len(out)} final recommendations")
        return out.to_dict(orient='records')
    
if 'data' in locals():
    print("üöÄ Initializing Hybrid Recommender System...")
    recommender = HybridRecommender(data)
else:
    print("‚ùå Please load data first")
```

```python
import numpy as np

def evaluate_content_model(content_model, test_ratings, k=10, rating_threshold=4.0):
    precisions, recalls, ndcgs = [], [], []

    # L·∫∑p qua t·ª´ng user
    for user_id in test_ratings['user_id'].unique():
        # L·∫•y c√°c m√≥n user ƒë√£ th√≠ch trong test set
        user_test = test_ratings[test_ratings['user_id'] == user_id]
        relevant_items = set(user_test[user_test['rating'] >= rating_threshold]['food_id'])
        if len(relevant_items) == 0:
            continue

        # Ch·ªçn 1 m√≥n user ƒë√£ th√≠ch trong train set ƒë·ªÉ l√†m "query item"
        liked_in_train = train_ratings[
            (train_ratings['user_id'] == user_id) &
            (train_ratings['rating'] >= rating_threshold)
        ]['food_id'].values
        if len(liked_in_train) == 0:
            continue

        query_item = liked_in_train[0]  # l·∫•y m√≥n ƒë·∫ßu ti√™n user th√≠ch trong train

        # G·ª£i √Ω m√≥n t∆∞∆°ng t·ª± t·ª´ content_model
        rec_indices, scores = content_model.recommend_similar_items(query_item, topk=k)
        recommended_items = set(data.recipes.iloc[rec_indices]['food_id'])

        # T√≠nh Precision@K v√† Recall@K
        hits = len(recommended_items & relevant_items)
        precisions.append(hits / k)
        recalls.append(hits / len(relevant_items))

        # T√≠nh NDCG@K
        recommended_list = list(data.recipes.iloc[rec_indices]['food_id'])
        dcg = 0.0
        for i, item in enumerate(recommended_list):
            if item in relevant_items:
                dcg += 1 / np.log2(i + 2)
        idcg = sum(1 / np.log2(i + 2) for i in range(min(len(relevant_items), k)))
        ndcgs.append(dcg / idcg if idcg > 0 else 0.0)

    return np.mean(precisions), np.mean(recalls), np.mean(ndcgs)
```

```python
p, r, ndcg = evaluate_content_model(content_model, test_ratings, k=10)
print(f"[Content] Precision@10={p:.4f}, Recall@10={r:.4f}, NDCG@10={ndcg:.4f}")
```

```python
def evaluate_collab_model(collab_model, test_ratings, train_ratings, k=10):
    """Evaluate collaborative model using Precision@K, Recall@K, and NDCG@K"""
    idx_to_recipe_id = {v: k for k, v in collab_model.recipe_id_to_idx.items()}
    grouped = test_ratings.groupby("user_id")
    
    precisions, recalls, ndcgs = [], [], []

    for user_id, group in grouped:
        # B·ªè qua user cold-start
        if user_id not in collab_model.user_id_to_idx:
            continue

        # C√°c m√≥n user th√≠ch trong test (relevant items)
        relevant_items = set(group[group["rating"] >= 3]["food_id"])
        if len(relevant_items) == 0:
            continue

        # L·∫•y g·ª£i √Ω t·ª´ collab_model
        rec_indices, scores = collab_model.recommend_for_user(user_id, topk=k)
        recommended_items = set(idx_to_recipe_id[idx] for idx in rec_indices)

        # Precision@K
        hit_count = len(recommended_items & relevant_items)
        precision = hit_count / k
        recall = hit_count / len(relevant_items)

        # NDCG@K
        dcg = 0.0
        idcg = sum(1.0 / np.log2(i + 2) for i in range(min(len(relevant_items), k)))
        for rank, item in enumerate(rec_indices[:k]):
            if idx_to_recipe_id[item] in relevant_items:
                dcg += 1.0 / np.log2(rank + 2)
        ndcg = dcg / idcg if idcg > 0 else 0.0

        precisions.append(precision)
        recalls.append(recall)
        ndcgs.append(ndcg)

    return (
        np.mean(precisions) if precisions else 0,
        np.mean(recalls) if recalls else 0,
        np.mean(ndcgs) if ndcgs else 0,
    )

# ƒê√°nh gi√° collaborative filtering
p, r, ndcg = evaluate_collab_model(collab_model, test_ratings, train_ratings, k=10)
print(f"[Collaborative] Precision@10={p:.4f}, Recall@10={r:.4f}, NDCG@10={ndcg:.4f}")
```

```python
def evaluate_hybrid_model(hybrid_model, test_ratings, k=10):
    """
    ƒê√°nh gi√° Hybrid Recommender model v·ªõi Precision@K, Recall@K, NDCG@K.
    """
    precisions, recalls, ndcgs = [], [], []
    user_groups = test_ratings.groupby("user_id")

    for user_id, group in user_groups:
        # L·∫•y m√≥n user th√≠ch trong test
        relevant_items = set(group[group["rating"] >= 3]["food_id"])
        if len(relevant_items) == 0:
            continue

        # G·ª£i √Ω top-k m√≥n ƒÉn t·ª´ hybrid_model
        recommended_items, _ = hybrid_model.recommend(user_id, k=k)
        recommended_items = set(recommended_items)

        # Precision@K
        hit_items = recommended_items & relevant_items
        precision = len(hit_items) / k
        precisions.append(precision)

        # Recall@K
        recall = len(hit_items) / len(relevant_items)
        recalls.append(recall)

        # NDCG@K
        dcg = sum([1 / np.log2(idx + 2) for idx, item in enumerate(recommended_items) if item in relevant_items])
        idcg = sum([1 / np.log2(idx + 2) for idx in range(min(len(relevant_items), k))])
        ndcg = dcg / idcg if idcg > 0 else 0
        ndcgs.append(ndcg)

    return np.mean(precisions), np.mean(recalls), np.mean(ndcgs)

# ---- G·ªçi ƒë√°nh gi√° ----
p, r, ndcg = evaluate_hybrid_model(recommender, test_ratings, k=10)
print(f"[Hybrid] Precision@10={p:.4f}, Recall@10={r:.4f}, NDCG@10={ndcg:.4f}")
```
